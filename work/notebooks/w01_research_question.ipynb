## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as  pd
df = pd.read_csv("/content/content_refresh_anonymized.csv")
total_pages= len(df)
declining_pages = (df["trend_direction"] == "down").sum()
declining_pct = (declining_pages / total_pages)* 100
avg_impression = df["impressions_90d"].mean()
print(f"Total pages: {total_pages}")
print(f"Declining pages: {declining_pages}")
print(f"Percentage declining: {declining_pct:.2f}%")
print(f"Average impressions (90d): {avg_impression:.2f}")
```

1. The starter dataset contains 30,000 content pages, providing a sufficiently large sample for analyzing content performance.
2. Among these pages, 16,262 (54.21%) are labeled as declining, indicating that more than half of the content shows signs of downward performance under the starter definition.
3. Additionally, the average number of impressions over the last 90 days is 5,200.37, suggesting that many pages have meaningful search visibility.
4. These statistics indicate that the dataset contains enough variation in content performance to build and evaluate a data-driven prioritization system for content review.
