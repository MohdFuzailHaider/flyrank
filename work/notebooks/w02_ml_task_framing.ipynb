## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

* Machine learning can learn patterns from the combination of these signals and assign a priority score that better reflects the overall likelihood that a page deserves review.
* This allows the ranking to adapt to more complex relationships that a page deserves review.
* This allows the ranking to adapt to more complex relationships than a simple rule-based approach while supporting human decision-making.
